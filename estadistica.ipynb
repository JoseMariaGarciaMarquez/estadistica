```python
import numpy as np
import pandas as pd
import scipy.stats as sts
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def estadistica(datos,x,y):
    #----Datos
    datos.head()
    #----Limpieza de datos
    resp = int(input("¬øDesea pre-procesar los datos?\n1. SI\n2. NO\n"))
    if resp == 1:
        repe = int(input("Eliga el pre-procesado que desea:\n1. DATOS REPETIDOS\n2. DATOS INCOMPLETOS\n"))
        if repe == 1:
            print("‚öôÔ∏èSU SET DE DATOS HAN SIDO PRE-PROCESADOS‚öôÔ∏è")
    print("-----------------------------\n"
        "üìäCALCULAR LOS ESTAD√çSTICOSüìä\n"
         "-----------------------------")
    medM = np.mean(datos)
    medianM = np.median(datos)
    modaM = sts.mode(datos)
    varM = np.var(datos)
    stdM = np.std(datos)
    asiM = sts.skew(datos)
    curM = sts.kurtosis(datos)

    print("TENDENCIA CENTRAL\nMedia: {:2.4f} ppm\nMediana: {:2.4f} ppm \nModa: {} ppm\n"
      "\nDISPERSION:\nVarianza: {:2.4f} ppp^2\nDesviacion estandar: {:2.4f} ppm\n"
      "\nFORMA:\nCoeficiente de asimetria: {:2.4f}\nCoeficiente de curtosis: {:2.4f}\n".format(medM, medianM, modaM[0][0], varM, stdM, asiM, curM))

    print("---------------------------\n"
        "üß≠üó∫GR√ÅFICOS EXPLORATORIOSüó∫üß≠\n"
         "---------------------------")
    fig, ax = plt.subplots(2,2, figsize = (8,8), dpi = 200)
    ax[0,0].hist(datos, color = 'darkorange')
    ax[1,0].boxplot(datos, labels = "D", vert = False, sym = '+r')
    ax[0,1].scatter(x,y, c = datos)
    ax[1,1].hist(datos, bins = 100, histtype = 'stepfilled', cumulative = True)
    plt.suptitle("GRAFICOS EXPLORATORIOS")
    plt.show()
    
    trans = int(input("¬øDesea transformar los datos?\n1. SI\n2. NO\n"))
    if trans == 1:
        op = int(input("Escoga una transformaci√≥n:\n1. RAIZ\n2. LOG\n3. Z-SCORE\n"))
        if op ==1:
            print("FUNCIONA RAIZ")
            datos = np.sqrt(datos)
            fig, ax = plt.subplots(2,2, figsize = (8,8), dpi = 200)
            ax[0,0].hist(datos, color = 'darkorange')
            ax[1,0].boxplot(datos, labels = "D", vert = False, sym = '+r')
            ax[0,1].scatter(x,y, c = datos)
            ax[1,1].hist(datos, bins = 100, histtype = 'stepfilled', cumulative = True)
            plt.suptitle("GRAFICOS EXPLORATORIOS")
            
        elif op == 2:
            print("FUNCIONA LOG")
            datos = np.log(datos)
            fig, ax = plt.subplots(2,2, figsize = (8,8), dpi = 200)
            ax[0,0].hist(datos, color = 'darkorange')
            ax[1,0].boxplot(datos, labels = "D", vert = False, sym = '+r')
            ax[0,1].scatter(x,y, c = datos)
            ax[1,1].hist(datos, bins = 100, histtype = 'stepfilled', cumulative = True)
            plt.suptitle("GRAFICOS EXPLORATORIOS")
        else:
            print("FUNCIONA Z")
            datos = sts.zscore(datos)
            fig, ax = plt.subplots(2,2, figsize = (8,8), dpi = 200)
            ax[0,0].hist(datos, color = 'darkorange')
            ax[1,0].boxplot(datos, labels = "D", vert = False, sym = '+r')
            ax[0,1].scatter(x,y, c = datos)
            ax[1,1].hist(datos, bins = 100, histtype = 'stepfilled', cumulative = True)
            plt.suptitle("GRAFICOS EXPLORATORIOS")
    else:
        
        fig, ax = plt.subplots(2,2, figsize = (8,8), dpi = 200)
        ax[0,0].hist(datos, color = 'darkorange')
        ax[1,0].boxplot(datos, labels = "D", vert = False, sym = '+r')
        ax[0,1].scatter(x,y, c = datos)
        ax[1,1].hist(datos, bins = 100, histtype = 'stepfilled', cumulative = True)
        plt.suptitle("GRAFICOS EXPLORATORIOS")
        plt.savefig("RESULTADOS.png")
        print("ü•ëResultados salvadosü•ë")
    
    return datos
    
        
```

```python
mu, sigma = 0,0.1;
x = np.arange(1000)
y = x;
datos = np.random.normal(mu, sigma,1000)
```

```python
estadistica(datos,x,y);
```
